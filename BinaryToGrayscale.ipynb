```python
import pandas as pd
import numpy as np
from PIL import Image
```

# **1. Convert bộ Benign sang Grayscale image (staDynBenignLab.csv)**

```python
# Load the binary CSV file
data_benign = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Grayscale-Image/staDynBenignLab.csv')
df_benign = data_benign.iloc[:,2:]
```

```python
df_benign = df_benign.astype(int)

# Convert numeric values to binary representation
binary_dataset = df_benign.applymap(lambda x: 1 if x > 0 else 0) # hàm x sẽ là 1 nếu x > 0 và ngược lại

print(binary_dataset['dd'])
```

```python
# Extract the label column
labels = binary_dataset['label'] # lấy côtk label

# Remove the label column from the data
binary_dataset = binary_dataset.drop('label', axis=1) # drop label
```

```python
# Convert the DataFrame to a NumPy array
data_array = binary_dataset.values
print(data_array.shape)
```

```python
data_benign['filename'].values
```

```python
filenames = data_benign['filename'].values # lấy tên file của bộ dl

# Convert the image data to grayscale images
grayscale_images = [] # tạo ra một mảng chứa ảnh
for i in range(data_array.shape[0]): 
    img_data = data_array[i]
    img_data *= 255 # chuyển thành trắng đen   0* 255 = 0, 1*255= 255
    img_data = img_data.astype(np.uint8)
    img = Image.fromarray(img_data, mode='L')
    img = img.resize((33, 33)) # lấy căn của số đặc tính
    grayscale_images.append(img)

# Save the grayscale images with corresponding labels
for i, img in enumerate(grayscale_images):
    filename = filenames[i]
    img.save(f'/content/drive/MyDrive/Colab Notebooks/Grayscale-Image/Benign-Images/{filename}.png')
```
